Más operaciones

```python
import numpy as np
a=float(input("ingrese el valor de a..."))
b=float(input("ingrese el valor de b..."))
c=a+b
d=a-b
e=a*b
f=a**2
# los dos asteriscos son elevar a
g=np. sqrt(b)
# los dos asteriscos y el un medio es raiz de a
h=np. log(b)
j=np. exp(a)
print("La suma tiene como resultado...",c)
print("La resta tiene como resultado...",d)
print("La multiplicación tiene como resultado...",e)
print("El resultado de la exponencia es... ",f)
print("El resultado de la raiz es...",g)
print("El logaritmo de b es...",h)
print("El exponente de a es... ",j)
```
