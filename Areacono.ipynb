Calcular el área de un cono de altura inclinada 15 cm y diámetro 6 cm.

```python
import numpy as np
Ainclinada=int(input("Ingrese la altura inclinada en cm..."));
D=int(input("Ingrese el diámetro en cm..."));
R=int(input("Ingrese el radio del cono en cm..."));
```

```python
A=(np.pi*R*(R+Ainclinada))
print("El área del cono es:...",A)
```
