Una locomotora necesita 10 s. para alcanzar su velocidad normal que es 25 m/s.
Suponiendo que su movimiento es uniformemente acelerado 

```python
Vinicial=int(input("La velocidad inical es de  m/s..."));
Vfinal=int(input("La velocidad final es de  m/s..."));
tiempo=int(input("El tiempo es de  s..."));
a=((Vfinal-Vinicial)/tiempo)
print("El resultado es...",Vfinal-Vinicial)
print("El resultado final es:...",a,"m/s2")
```

```python
Vinicial=float(input("Ingrese la velocidad inicial en m/s2"))
tiempo=float(input("Ingrese el tiempo en s"))
a=float(input("Ingrese la aceleración en m/s:..."))
d=(Vinicial*tiempo+1/2)*(a*(tiempo**2))
print("La distancia es:...",d,"m")
```
