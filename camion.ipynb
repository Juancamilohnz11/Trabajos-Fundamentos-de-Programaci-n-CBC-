Un camión circula por una carretea a 20m/s . En 5 s , su velocidad pasa a ser de 25 m/s ¿ cuál ha sido su aceleración ?

```python
Vinicial=int(input("La velocidad inical es de  m/s..."));
Vfinal=int(input("La velocidad final es de  m/s..."));
tiempo=int(input("El tiempo es de  s..."));
```

```python
Aceleración=int((Vfinal-Vinicial)/tiempo)
```

```python
print("La aceleración del camión es de m/s2",Aceleración);
```
